#### Graph ploting by using seaborn and matplotlib:

```python
plt.plot(df['brand'])
plt.ylabel("Name of brand")
plt.show()
```

```python
sns.lineplot(df["price"])
plt.show()
```

```python
sns.scatterplot(x = "power",y = "price",hue = 'owner',data = df)
plt.title("Age vs Price")
plt.show()
```

```python
plt.hist(df['owner'])
plt.show()
```

```python
sns.histplot(x = 'brand', kde = True,data = df)
plt.title("Bike brand histogram")
plt.xticks(rotation = 80)
plt.show()
```

```python
sns.kdeplot(df['power'])
plt.show()
```

```python
sns.relplot(x = 'price',y = "age",hue = "owner",data = df)
plt.title("Relation between price and age")
plt.xlabel('price',fontsize = 20,color = "red")
plt.ylabel("age")
plt.show()
```

```python
sns.countplot(x = "age",data = df)
plt.title("Age count graph")
plt.xticks(rotation  = 90)
plt.show()
```

```python
sns.countplot(x = 'owner',data=df)
plt.title("Number of owner ")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x = 'owner',y = 'price',hue = "owner",kind = 'bar',data=df)
plt.title("Price Distribution of owner")
plt.xticks(rotation=50)
plt.show()
```

```python
sns.catplot(x = 'owner',y = 'price',kind = 'violin',data=df)
plt.title("Price Distribution of owner")
plt.xticks(rotation=50)
plt.show()
```

```python
ratings = df.groupby('owner')[[
    "price",
    "kms_driven",
    "age",
    "power"
]].mean()

ratings.plot(kind = "bar")
plt.title("Average owner rating ")
plt.xticks(rotation = 70)
plt.show()
```

```python
brand_count = df['owner'].value_counts()

plt.pie(
    brand_count.values,
        labels = brand_count.index,
        explode=[0,0,0,2],
        shadow =True,
        autopct='%1.1f%%')
plt.title("Brand Distribution")
plt.show()
```
